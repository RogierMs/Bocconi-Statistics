### Exercise 3

```python
def lag(x, n):
    x = np.delete(x, np.s_[len(x)-n:])
    x = np.append(x, np.repeat(np.NaN, n))
    return x
```

```python
lag([1,2,3,4,5], 2)
```

```python
def lead(x, n):
    x = np.array(x, dtype=np.float) #convert values in array so that NaN can be inserted
    x = np.delete(x, np.s_[:n])
    x = np.insert(x, 0, np.repeat(np.NaN, n))
    return x
```

```python
lead([1,2,3,4,5], 2)
```

### Exercise 1

```python
data_weather = pd.read_csv("nycflights13_weather.csv", sep=',',skiprows=42, header=0)
```

```python
data_weather.head()
```

### Temperature Conversion

```python
#F = 1.8C + 32 => C = (F - 32)/1.8
```

```python
data_weather['temp'] = np.round((data_weather['temp'] - 32)/1.8, 4)
```

```python
data_weather.head()
```

### Daily Mean Temperatures for the JFK Airport

```python
data_weather["date"] = data_weather["year"].map(str) + data_weather["month"].map(str) + data_weather["day"].map(str)
```

```python
data_weather["date"] = data_weather["date"].map(int) #convert to int for sorting purposes
```

```python
data_weather.head()
```

```python
jfk_avg_temp = data_weather[["temp", "date"]][data_weather["origin"] == "JFK"].groupby("date").mean()
```

```python
jfk_avg_temp.head()
```

### Greater mean temperature than on the preceding day

```python
avg_temp = data_weather[["temp", "date"]].groupby("date").mean().reset_index()
```

```python
avg_temp.head()
```

```python
for row in range(1,len(avg_temp)):
    if avg_temp["temp"][row] > avg_temp["temp"][row-1]:
        print(avg_temp["date"][row])
```

```python
avg_temp["temp"][1] > avg_temp["temp"][0]
```

###  5 Hottest days

```python
avg_temp = data_weather[["temp", "date"]].groupby("date").mean()
```

```python
hottest_days = avg_temp["temp"].nlargest(5)
```

```python
hottest_days
```
